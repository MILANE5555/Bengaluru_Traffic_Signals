```python
osm = esy.osm.pbf.File('/content/drive/MyDrive/bengaluru.osm.pbf')
```

```python
tf_summary = []
tf_all_details = []
for entry in osm:
  if(len(entry.tags) != 0):
    for x in entry.tags:
      if(len(x) != 0 and entry.tags[x] == 'traffic_signals'):
        dict = {}
        dict['id'] = entry.id
        dict['lon'] = entry.lonlat[0]
        dict['lat'] = entry.lonlat[1]
        tf_summary.append(dict)
        tf_all_details.append(entry)
        #print(entry)
    
```
